```python
import pandas as pd
import sqlite3

df = pd.read_csv(r"C:\Users\Dell\train.csv")
conn = sqlite3.connect("superstore.db")
df.to_sql("orders", conn, if_exists="replace",index=False)
```

```python
pd.read_sql("SELECT * FROM orders LIMIT 10;",conn)
```

```python
pd.read_sql("SELECT * FROM orders WHERE Region = 'West';",conn)
```

```python
# Rename columns: lowercase + underscores


df.columns = [col.strip().lower().replace(" ","_") for col in df.columns]


df.to_sql("orders",conn, if_exists="replace",index=False)

print(df.columns)
```
